### 1(c)ii. Extract the time-domain features minimum, maximum, mean, median, standard deviation, first quartile, and third quartile for all of the 6 time series in each instance. You are free to normalize/standardize features or use them directly.

```python
def calc_summary(dframe,summary):
    column = dframe[0].columns
    column=column[1:]
    for i in range(len(dframe)): #iterates through all instances
        curr_df=dframe[i] #current instance as a dataframe
        #print(curr_df)
        cuur_df=curr_df.drop(columns=['time'])
    
        new_row=list() #details for each instance is recorded in this list and added to the summary dataframe
        new_row.append(int(i+1)) #appends the instance number
    
        for col in column:
            described = curr_df[col].describe()
            new_row.append(described["min"])
            new_row.append(described["max"])
            new_row.append(described["mean"])
            new_row.append(described["50%"])
            new_row.append(described["std"])
            new_row.append(described["25%"])
            new_row.append(described["75%"])
        #print(new_row,'\n')
        summary_length = len(summary)
        summary.loc[summary_length] = new_row #appending statistical summary for new instance at the end of the dataframe
    
    #converting the instance column from float to integer   
    convert= {'instance': int } 
    summary = summary.astype(convert) 
    return summary
ans_df=pd.DataFrame(columns=['instance','min1','max1','mean1','median1','SD1','1st_Quart1','3rd_Quart1',
                         'min2','max2','mean2','median2','SD2','1st_Quart2','3rd_Quart2',
                         'min3','max3','mean3','median3','SD3','1st_Quart3','3rd_Quart3',
                         'min4','max4','mean4','median4','SD4','1st_Quart4','3rd_Quart4',
                         'min5','max5','mean5','median5','SD5','1st_Quart5','3rd_Quart5',
                         'min6','max6','mean6','median6','SD6','1st_Quart6','3rd_Quart6'
                        ])
ans=calc_summary(alldata,ans_df)
ans
```

### 1(c)iii. Estimate the standard deviation of each of the time-domain features you extracted from the data. Then, use Python’s bootstrapped or any other method to build a 90% bootsrap confidence interval for the standard deviation of each feature.

```python
#The standard deviation for all the domain features are
feature_std=dict()
ans=ans[['min1','max1','mean1','median1','SD1','1st_Quart1','3rd_Quart1',
                         'min2','max2','mean2','median2','SD2','1st_Quart2','3rd_Quart2',
                         'min3','max3','mean3','median3','SD3','1st_Quart3','3rd_Quart3',
                         'min4','max4','mean4','median4','SD4','1st_Quart4','3rd_Quart4',
                         'min5','max5','mean5','median5','SD5','1st_Quart5','3rd_Quart5',
                         'min6','max6','mean6','median6','SD6','1st_Quart6','3rd_Quart6'
                        ]]
for col in ans.columns:
    described = ans[col].describe()
    feature_std[col]=described["std"]

for k , v in feature_std.items(): 
    print( k+"\t", v)
```

```python
#Bootstrap confidence interval for the standard deviation of each feature
#Using Python's bootstrapped

#reference: https://pypi.org/project/bootstrapped/
import bootstrapped.bootstrap as bs
import bootstrapped.stats_functions as bs_stats


cols=['min1','max1','mean1','median1','SD1','1st_Quart1','3rd_Quart1',
                         'min2','max2','mean2','median2','SD2','1st_Quart2','3rd_Quart2',
                         'min3','max3','mean3','median3','SD3','1st_Quart3','3rd_Quart3',
                         'min4','max4','mean4','median4','SD4','1st_Quart4','3rd_Quart4',
                         'min5','max5','mean5','median5','SD5','1st_Quart5','3rd_Quart5',
                         'min6','max6','mean6','median6','SD6','1st_Quart6','3rd_Quart6'
                        ]
CI=list()
for c in cols:
    coldata=np.array(ans[c])
    currCI=bs.bootstrap(coldata, stat_func=bs_stats.std, alpha=0.1)
    #print(currCI)
    CI.append([c,currCI.value,currCI.lower_bound,currCI.upper_bound])
   
CI_df = pd.DataFrame(CI, columns=["Feature","Value","Lower Bound","Upper Bound"])
print("90% Bootsrap confidence interval for the standard deviation of each feature is as follows-:")
CI_df
```

### 1(d)iii. Break each time series in your training set into l ∈ {1, 2, . . . , 20} time series of approximately equal length and use logistic regression to solve the binary classification problem, using time-domain features. Remember that breaking each of the time series does not change the number of instances. It only changes the number of features for each instance. Calculate the p-values for your logistic regression parameters in each model corresponding to each value of l and refit a logistic regression model using your pruned set of features. Alternatively, you can use backward selection using sklearn.feature selection or glm in R. Use 5-fold cross-validation to determine the best value of the pair (l, p), where p is the number of features used in recursive feature elimination.

```python
#function to break time series in l parts
def SplitData(dflist,divide):
    
    # new data column list
    new_cols = []
    for i in range(1,6*divide+1):
        new_cols.extend(["max"+str(i),"mean"+str(i),"median"+str(i)])
        #new_cols.extend(["min"+str(i),"max"+str(i),"mean"+str(i),"median"+str(i),"SD"+str(i),"1st_Quart"+str(i),"3rd_Quart"+str(i)])
    #print(len(new_cols))
    column = ["avg_rss12", "var_rss12", "avg_rss13", "var_rss13","vg_rss23", "ar_rss23"]

    datalist = []
    for curr_df in dflist:
        location=[0]
        num = len(curr_df)
        dn = divide
        for l in range(divide):
            result = int(num/dn)
            #print(result)
            location.append(location[l]+result)
            dn = dn-1
            num = num - result
            
        #print(location)
        #for d in range(divide):
            #print(location[d])
            #print(location[d+1])
        dslist = [curr_df.iloc[location[d]:location[d+1]] for d in range(divide)]
        #print(dslist)
        new_row=list() #details for each instance is recorded in this list and added to the datalist
        for ds in dslist:
            #print(ds,'\n\n\n\n')
            for col in column:
                described = ds[col].describe()
                #new_row.append(described["min"])
                new_row.append(described["max"])
                new_row.append(described["mean"])
                new_row.append(described["50%"])
                #new_row.append(described["std"])
                #new_row.append(described["25%"])
                #new_row.append(described["75%"])
        #print(len(new_row),'\n')
        datalist.append(new_row)

        
    #print(result_df)    
    return pd.DataFrame(datalist, columns=new_cols)

#tryone=SplitData(trainlist,7)
#tryone
```

```python
from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score, confusion_matrix, roc_curve, auc
import statsmodels.api as sm
import math

L = np.arange(1,21,1)
y=[1 if x=='bending1' or x=='bending2' else 0 for x in train_y]
yy=[1 if x=='bending1' or x=='bending2' else 0 for x in test_y]
    
bestlp = list()

#%% 
# Calculate RFE
lp_score = dict()
best_features_df = dict()
for l in L:
    X = SplitData(trainlist,l)
    print(l,'\n')
    Y = y
    reg = LogisticRegression(C=math.inf, max_iter=1000,n_jobs=-1) #To remove regularization
    rfecv = RFECV(estimator=reg, cv=StratifiedKFold(5), scoring='accuracy')
    rfecv=rfecv.fit(X,Y)
    #gives the cv score
    lp_score[l] = rfecv.grid_scores_[rfecv.n_features_ - 1]
    feature = list()
    for index, col in enumerate(X.columns):
        if rfecv.support_[index] == True:
           feature.append(col) 
    best_features_df[l] = feature

```

```python
# Show the best (l,p)
print("Results based on Accuracy")
for key, value in lp_score.items():
    print('{key}:{value}'.format(key=key, value=value))
print('Best L is', max(lp_score, key=lp_score.get), '\n', ' and corresponding p is:', len(best_features_df[max(lp_score, key=lp_score.get)]))
```

```python
best_l=max(lp_score, key=lp_score.get)
selectedFeatures=best_features_df[best_l]
best_p=len(selectedFeatures)
print("The selected features are ",selectedFeatures)
```

#### Explain what the right way and the wrong way are to perform cross-validation in this problem.

##### Wrong way
Do feature selection and then applying cross-validation(split data into train and validation sets) to selected variables

##### Right way
First split the data into train and validation sets and do feature selection based on the train set.

##### Why? 
If we choose the 'right way', we would have multiple variables choices which would be better because there is variability in the actual population. If we go back to the original population and take another training set, would we get the same set of highest correlation variables? Most probably no. We should take into consideration this variability. In other words, feature selection should be included in cross validation, not outside the cv process. By doing the wrong way, the model we built in feature selection has already seen the validation data, which turns out to have a good performance in predicting validation data.

```python
#Reference code for calculating the p-values
class LogisticReg:
    def __init__(self,*args,**kwargs):#,**kwargs):
        self.model = linear_model.LogisticRegression(*args,**kwargs)#,**args)

    def fit(self,X,y):
        self.model.fit(X,y)
        #### Get p-values for the fitted model ####
        denom = (2.0*(1.0+np.cosh(self.model.decision_function(X))))
        denom = np.tile(denom,(X.shape[1],1)).T
        F_ij = np.dot((X/denom).T,X) ## Fisher Information Matrix
        Cramer_Rao = np.linalg.inv(F_ij) ## Inverse Information Matrix
        sigma_estimates = np.sqrt(np.diagonal(Cramer_Rao))
        z_scores = self.model.coef_[0]/sigma_estimates # z-score for eaach model coefficient
        p_values = [stat.norm.sf(abs(x))*2 for x in z_scores] ### two tailed test for p-values
        
        self.z_scores = z_scores
        self.p_values = p_values
        self.sigma_estimates = sigma_estimates
        self.F_ij = F_ij
        self.coef_ = self.model.coef_[0]
        return self.coef_,self.sigma_estimates,self.z_scores,self.p_values
        
    def predict(self,X):
        return self.model.predict(X)
    
    def score(self,X,y):
        return self.model.score(X,y)

LRmodelpv = LogisticReg(C=math.inf,n_jobs=-1)
coefs, se, z, p = LRmodelpv.fit(Xtrain,Ytrain)
prediction = LRmodelpv.predict(Xtrain)
confM = confusion_matrix(Ytrain,prediction)
confM
t=list()
for x in range(len(Xtrain.columns)):
    t.append([Xtrain.columns[x],round(coefs[x],4),round(se[x],6),round(z[x],6),round(p[x],5)])
dff=pd.DataFrame(t,columns=["Predictor","Coefficient","SE","Z-statistic","p-value"])
dff

```

### 1(e) Binary Classification Using L1-penalized logistic regression
### i. Repeat 1(d)iii using L1-penalized logistic regression, i.e. instead of using p- values for variable selection, use L1 regularization. Note that in this problem, you have to cross-validate for both l, the number of time series into which you break each of your instances, and λ, the weight of L1 penalty in your logistic regression objective function (or C, the budget). Packages usually perform cross-validation for λ automatically.

### (a) Suppose that the true relationship between X and Y is linear, i.e. Y = β0 + β1X + ε. Consider the training residual sum of squares (RSS) for the linear regression, and also the training RSS for the cubic regression. Would we expect one to be lower than the other, would we expect them to be the same, or is there not enough information to tell? Justify your answer.


A cubic regression model would be more flexible(i.e, more wigly and would curve to fit more points than a straight line) and one would expect the RSS in the training data to be smaller when compared to a linear regression model. As we are assuming that the true relationship between X and Y is linear, a cubic regression model overfit the 100 training points in the specific training set,even fitting the noise in the data. This would cause the RSS for the cubic model to be lesser than the linear model.

### (b) Answer (a) using test rather than training RSS.

We would expect that the cubic regression model would fit the noise of training data, leading to lower training RSS than a linear model but at the expense of less accurate predictive capability.
It is a bit difficult to say,but for test data, I would expect the RSS to be smaller for linear regression as opposed to the cubic regression because the parameters of cubic regression might be wrong due to overfitting of training data, and it would miss to capture the actual underlying relationship between the predictor and the response, which the linear might be successfully be able to.





### (c) Suppose that the true relationship between X and Y is not linear, but we don’t know how far it is from linear. Consider the training RSS for the linear regression, and also the training RSS for the cubic regression. Would we expect one to be lower than the other, would we expect them to be the same, or is there not enough information to tell? Justify your answer.

Assuming that the true relationship between X and Y was non-linear, I would expect the cubic regression model to fit the training data better due to its increased flexibility and would capture the underlying relationship pattern better(when compared to a straight line which a linear model would generate). My reasoning is the same as in part (a), but in this case, I believe it is even more likely that the increased flexibility of cubic regression model would lead to a lower RSS.

### (d) Answer (c) using test rather than training RSS.

It is a bit difficult to answer which model would have a lower test RSS. We don't know how far exactly the relationship of X and Y is from linear. If the true relationship is far from linear (and especially if it happens to be cubic), cubic regression will lead to lower RSS in test data as it is a better representation of the underlying data. On the contrary, if the relationship between X and Y is very close to linear, linear regression will have lower RSS in the test data. The explanation is similar to part 2(b).
